```python
word = 'abcde'

for item in enumerate(word):
    print(item)
```

```python
for x, y in enumerate(word):
    print(x)
    print(y)
```

```python
for x, y in enumerate(word):
    print(x)
    print(f"'{y}'")
```

```python
word = "Besant"

#Use tuple and packing
for index, letter in enumerate(word):
    print(index)
    print(letter)
    print("\n")
```
