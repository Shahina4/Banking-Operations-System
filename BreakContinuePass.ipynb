break – Stops the loop entirely.

Exits the loop even if the condition is still true. Example,

for i in range(5):
    if i == 3:
        break
    print(i)
# Output: 0 1 2

continue – Skips the current iteration and moves to the next one.

Doesn't exit the loop; just skips to the next iteration. Example,

for i in range(5):
    if i == 3:
        continue
    print(i)
# Output: 0 1 2 4

pass – Does nothing; a placeholder.

Used when a statement is required syntactically but no action is needed. Example,

for i in range(5):
    if i == 3:
        pass
    print(i)
# Output: 0 1 2 3 4
