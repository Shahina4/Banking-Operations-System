```python
#class Fruit():
class Car():
    tyres = 4
    def __init__(self):
        print("This is Car class")
    def engine(self):
        print("The Car has a powerful engine")
    def doors(self):
        print("This Car has four doors")

class Swift(Car):
    def __init__(self):
        Car.__init__(self)
        print("Swift Initialized")
    def sunRoof(self):
        print("The new Swift has sun roof as per Manoj")
```

```python
obj = Swift()
```

```python
obj.engine()
obj.doors()
obj.sunRoof()
```

```python
#class Fruit():
class Car():
    tyres = 4
    def __init__(self):
        print("This is Car class")
    def engine(self):
        print("The Car has a powerful engine")
    def doors(self):
        print("This Car has four doors")

class Swift(Car):
    def __init__(self):
        Car.__init__(self)
        print("Swift Initialized")
    def sunRoof(self):
        print("The new Swift has sun roof as per Manoj")
    def engine(self):    #method overriding is achieved using inheritance by defining the same method name with same no.of parameters in the child class
        print("The Car has an average engine")
```

```python
obj = Swift()
```

```python
obj.engine()
obj.doors()
obj.sunRoof()
```

```python
class Car():
    tyres = 4
    def __init__(self):
        print("This is Car class")
    def engine(self):
        print("The Car has a powerful engine")
    def doors(self):
        print("This Car has four doors")
    def brake(self):
        print("This Car has break feature")

class Tata(Car):
    def __init__(self):
        Car.__init__(self)
        print("Tata initialized")
    def ac(self):
        print("The new Tata has AC")
    def sunRoof(self):
        print("The new Tata has Sun Roof")
    def map(self):
        print("The new Tata has map feature")
```
