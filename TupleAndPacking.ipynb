```python
myList = [(1,2),(3,4),(5,6),(7,8)]
```

```python
len(myList)
```

```python
for (x,y) in myList:
    print(x)
```

```python
for (x,y) in myList:
    print(y)
```

```python
for (x,y) in myList:
    print(x)
    print(y)
```

```python
#Tuple and unpacking => by accessing tuple through variables for each item

work_hours = [('Bob',1000),('Miller',800), ('Chris',1200)]
def employee_check(work_hours):
    current_max = 0
    employee_of_month=''
    for employee, hours in work_hours:
        if hours>current_max:
            current_max = hours
            employee_of_month = employee
        else:
            pass
    return (employee_of_month,current_max)
```

```python
employee_check(work_hours)
```
