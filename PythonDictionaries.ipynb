Dictionaries are unordered mapping for storing objects. Previously we saw how lists store objects in an ordered sequence, 
dictionaries use key-value pairing instead.

This key-value pair allows user to quickly grab objects without needing to know an index location.

Dictionaries use curly braces and colons to signify the keys and their associated values
{'key1':'value1', 'key2':'value2'}

Dictionaries are MUTABLE

#Methods: keys(), values(), items(), type(), len(), upper(), lower()

Dictionary vs List:
    
Dictionaries are used when we don't know the index values and where the items cannot be sorted. We just need to know the key
value pair.

Lists can be retrieved based on location. Ordered sequence can be indexed or sliced.

KEYS SHOULD ALWAYS BE Strings. Values can be of ANY DATA TYPE.
