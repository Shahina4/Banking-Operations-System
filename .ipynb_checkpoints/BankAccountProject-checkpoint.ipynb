```python
class Account:
    def __init__(self, acc_number, acc_holder):
        self.acc_number = acc_number
        self.acc_holder = acc_holder
        self.balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited to {self.acc_number}. New balance: ₹{self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"₹{amount} withdrawn from {self.acc_number}. New balance: ₹{self.balance}")
        else:
            print("Insufficient balance or invalid amount.")

    def display_balance(self):
        print(f"Account {self.acc_number} | Holder: {self.acc_holder} | Balance: ₹{self.balance}")


class SavingsAccount(Account):
    def __init__(self, acc_number, acc_holder, interest_rate=0.03):
        super().__init__(acc_number, acc_holder)
        self.interest_rate = interest_rate
        print(f"Savings Account opened for {acc_holder} ({acc_number})")

    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f"Interest ₹{interest:.2f} added to {self.acc_number}. New balance: ₹{self.balance:.2f}")


class CurrentAccount(Account):
    def __init__(self, acc_number, acc_holder):
        super().__init__(acc_number, acc_holder)
        self.overdraft_limit = 0
        print(f"Current Account opened for {acc_holder} ({acc_number})")

    def set_overdraft_limit(self, limit):
        self.overdraft_limit = limit
        print(f"Overdraft limit of ₹{limit} set for {self.acc_number}")

    def withdraw(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f"₹{amount} withdrawn from {self.acc_number}. New balance: ₹{self.balance}")
        else:
            print(f"Overdraft limit exceeded for {self.acc_number}. Transaction denied.")


class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self, acc_type, acc_number, acc_holder):
        if acc_number in self.accounts:
            print("Account already exists!")
            return

        if acc_type.lower() == 'savings':
            self.accounts[acc_number] = SavingsAccount(acc_number, acc_holder)
        elif acc_type.lower() == 'current':
            self.accounts[acc_number] = CurrentAccount(acc_number, acc_holder)
        else:
            print("Invalid account type. Please choose 'savings' or 'current'.")
            return

        print(f"{acc_type.capitalize()} account {acc_number} created for {acc_holder}.")

    def get_account(self, acc_number):
        return self.accounts.get(acc_number, None)


# === Using the Enhanced Bank Application ===

bank = Bank()

# Create Accounts
bank.create_account('current', 'CA2010', 'Shahina')
bank.create_account('savings', 'SA1020', 'Ali')

# Current account actions
ca = bank.get_account('CA2010')
ca.set_overdraft_limit(500)
ca.deposit(1000)
ca.withdraw(1300)
ca.display_balance()

# Savings account actions
sa = bank.get_account('SA1020')
sa.deposit(2000)
sa.add_interest()
sa.display_balance()
```
