> Regular Expressions (regex) allow us to search for general patterns in text data!
 
> Email: user@gmail.com :: pattern - (r"text"+"@"+"text"+".com")
 
> Phone: (555)-555-5555 :: pattern - r"(\d\d\d)-\d\d\d-\d\d\d\d"

                         > pattern - r"(\d{3})-\d{3}-\d{4}"
                         
> Indian Phone: +91-55555-55555 > pattern - r"+91-\d{5}-\d{5}"

> The re library allows us to create specialized pattern strings and then search for matches within text.

> The primary skill set for regex is understanding the special syntax for these pattern strings.

```python
text = "The agent's phone number is 315-555-5656."
```

```python
'phone' in text
```

```python
import re
```

```python
pattern = 'phone'
```

```python
re.search(pattern, text)
```

```python
pattern = 'Not in text'
```

```python
re.search(pattern, text) #not found: return nothing 
```

```python
pattern = 'phone'
```

```python
match = re.search(pattern,text)
```

```python
match
```

```python
match.span()
```

```python
match.start()
```

```python
match.end()
```

```python
text = 'my phone once, my phone twice'
```

```python
match = re.search('phone', text)
```

```python
match
```

```python
matches = re.findall('phone', text)
```

```python
matches
```

```python
len(matches)
```

```python
for match in re.finditer('phone', text):
    print(match)
    print(match.span())
```
