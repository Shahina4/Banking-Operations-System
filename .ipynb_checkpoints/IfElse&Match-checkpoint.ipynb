One set of if else can have multiple elif statements.

```python
var = 30
print('I think I am wrong')
if var==40:
    print(var, 'is equal to 30')
elif var==20:
    print(var, 'is greater than 20')
elif var<25:
    print(var, 'is less than 2')
else:
    print('Nothing will happen!')
print('I am wrong!!')
```

python 3.10 after that switch case is there i.e, match case

```python

dayNo = 1

match dayNo:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    
```

```python

dayNo = 8

match dayNo:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid Day of a Week")
   
```
