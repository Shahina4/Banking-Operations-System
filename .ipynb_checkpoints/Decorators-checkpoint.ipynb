# Decorators

Decorators allow you to "decorate" afunction.

Definition: A decorator is a function that takes another function and extends the behavior of thet latter function without explicitly modifying it.

Case 1: Imagine we created a function:

            def simple_func():
                #Do simple stuff
                return something


Case 2: Now we want to add some new capabilities to the function:

            def simple_func():
                #Want to add more stuff!
                #Do simple stuff (As done previously)
                return something



While doing the above, we have 2 ptions:

    > Add that extra code to old existing function.
                        OR
    > Create a brand new function that contains the old code and then add new code to that.

In case the functionality needs to be removed, we would have to do it manually. However, there could also be an option to on and off the functionality - Decorators!


    Python has decorators that allows us to induce on extra functionality to an already existing function.

        The @ operator is used and are then placed on top of the original function.

        Use a decorator when there is a need to change the behavior of a function without modifying the function itself. Best exapmle is -

```python
def hello(name = 'Besant'):
    print('This is the start of hello()')

    def greet():
        return '\t This is the greet() function inside the hello!' #\t denotes tab space to show indentation.
    def welcome():
        return '\t This is the welcome() function inside hello'

    print(greet())
    print(welcome())
    print('This is the end of hello()')
```

```python
hello()
```

```python
welcome()
```

```python
def hello(name = 'Besant'):
    print("The hello() function has been executed")

    def greet():
        return '\t This is the greet() function inside the hello!' #\t denotes tab space to show indentation
    
    def welcome():
        return '\t This is the welcome() function inside hello'

    print('function is being returned!')

    if name == 'Besant':
        return greet
    else:
        return welcome()
```

```python
hello('Besantt')
```

```python
def new_decorator(original_fun):
    def wrap_func():
        print('Some extra code, before the original function')

        original_fun()

        print('Some extra code, after the original function')

    return wrap_func()
```

```python
def func_needs_decorator():
    print('I want to be decorated!')
```

```python
new_decorator(func_needs_decorator)
```

```python
func_needs_decorator()
```

```python
decorated_func = new_decorator(func_needs_decorator)
```

```python
type(decorated_func)
```

```python
decorated_func()
```

```python
@new_decorator
def new_function():
    print("A fresh function is just created!")
```

```python
new_function()
```
