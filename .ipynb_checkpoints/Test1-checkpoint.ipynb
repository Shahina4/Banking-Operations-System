```python
number = input("Enter mobile number:")
print("Given Number: {}".format(number))
reverse = ['' + number[i] for i in range(9,-1,-1)]
print("Reversed number:", ''.join(reverse))
odd = [reverse[i] for i in range(0,10) if int(reverse[i])%2 != 0]
print('Odd numbers:',''.join(odd))
odd.sort()
print('Odd numbers in ascending order:',''.join(odd))
```
