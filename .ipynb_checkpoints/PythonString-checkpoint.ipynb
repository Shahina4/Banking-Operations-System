split(): splitting a string into collection of items(in the form of list) based on whitespaces in a string

```python
print("I visit Besant Technologies to learn Python".split())   #splitting items 
```

```python
print("I visit Besant Technologees to learn Python".split('e'))
```

```python
print("I visit Besant Technologies     to learn Python".split()) 
#By default,is splitting string based on whitespace into list
#(without considering how many no.of whitespaces are there) when no args given 
```

```python
print("I visit Besant Technologies     to learn Python".split(' ')) #considers every whitespace and splits it as an item('')
#If we manually give ',',' ','_','e' whenever there are consecutive specified character(passed arg in split) in a string
#Then it will consider them as one item with empty string ''
#Ex: ("Hi,,,,Besant,Technologies".split(',')) => ['Hi', '', '', '', 'Besant', 'Technologies'] 
```

```python
print("I,visit,Besant,,,Technologies,to learn Python".split(','))
```
