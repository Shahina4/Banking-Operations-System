```python
Polymorphism:
    Polymorphism means many forms
> It allows objects of different classes to be treated as object of a common super class.
> It enables flexibility and extensibility in code.
```

```python
class Dog():
    def __init__(self, name):
        self.name = name
    def speak(self):
        return self.name + ' says Woof!'
class Cat():
    def __init__(self, name):
        self.name = name
    def speak(self):
        return self.name + ' says Meow!'
obj1 = Dog('Johnny')
obj2 = Cat('Tom')

print(obj1.speak())
print(obj2.speak())
```

```python
#Demonstrating polymorphism using for loop => used for each object
for pet in [obj1, obj2]:
    print(pet.speak())
```

```python
#Demonstrating polymorphism using functions => Used depends upon need or requirement
def petSpeak(pet):
    print(pet.speak())
petSpeak(obj1)
petSpeak(obj2)
```
