Special Methods:
    Classes in Python can implement certain operations with special method names. These methods are not actually called directly
but by Python specific language syntax.

```python
class Book():
    def __init__(self, title, author, pages):
        print('A book is created')
        self.title = title
        self.author = author
        self.pages = pages
    def __str__(self):
        return "Title: %s, Author: %s, Pages: %s" % (self.title, self.author, self.pages)
    def __len__(self):
        return self.pages
    def __del__(self):     #del is inbuilt special method which deletes an object
        print('This object is going to delete')
        return 'A book is destroyed'
        
pythonBook = Book('Python is Amazing', 'Besant Technologies', 160)
```

```python
class Demo:
    def __init__(self):
        print("Constructor: Object created.")

    def __del__(self):
        print("Destructor: Object destroyed.")

obj = Demo()
del obj  # Forces call to __del__()
```
