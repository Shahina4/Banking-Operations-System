```python
#30/5/2025
#Converting meters into kilometers => 1km = 1000 meters =>1meter = 0.001kilometers
meters = [1000, 2000, 15000, 16500]
kilometers = []
for m in meters:
    km=m/1000
    kilometers.append(km)
kilometers
```

```python
meters = [1000, 2000, 15000, 16500]
kilometers = [m/1000 for m in meters]
kilometers
```

```python
meters = [1000, 2000, 15000, 16500]
kilometers = [int(m/1000) for m in meters]
kilometers
```

0 or 1 = Bit
8 bits = 1 Byte
1024 Bytes = 1 KB
1024 KB = 1 MB
1024 MB = 1 GB
1024 GB = 1 TB

```python
#Converting bytes to MegaBytes => MB = 1024 KB = 1024 * 1024 Bytes 
Bytes = [675432244565, 4342558, 76545446, 234457867564]
megaBytes = [b/(1024*2) for b in Bytes]
megaBytes
```

```python
#Converting bytes to MegaBytes => MB = 1024 KB = 1024 * 1024 Bytes 
Bytes = [675432244565, 4342558, 76545446, 234457867564]
megaBytes = [int(b/(1024*2)) for b in Bytes]
megaBytes
```

```python
Bytes = [675432244565, 4342558, 76545446, 234457867564]
megaBytes = []
for b in Bytes:
    mb = round(b/(1024*2))
    megaBytes.append(mb)
megaBytes
```
