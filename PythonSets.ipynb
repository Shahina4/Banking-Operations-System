Sets are unordered collections of unique elements
Meaning there can be one representative of the same object
Sets don't have a particular order (sets are inbuilt with sorted technique in itself)
It is mutable
#type(), add(),pop() with no args: removes first item, len()

```python
mySet=set()
```

```python
type(mySet)
```

```python
mySet.add(100)
```

```python
mySet
```

```python
mySet.add(250)
```

```python
mySet
```

```python
mySet.add(250) #ignore repeated values
```

```python
mySet
```

If we try to add same value again to set, it will not throw an error nor will it add the value again. It will just consider
to ignore adding the repeated/duplicated values as it already exists in the set.
