# Python Debugger(pydb):
Generally, in order to figure out what errors are there in the code we have used the print statement to try to track that error. However, Python comes with a built-in debugger tool that allows to interactively explore variables within mid operation of the Python code.

```python
x = [1, 2, 3]
y = 2
z = 3

result = y + z
print(result)
result2 = x + y
print(y)
print(result2)
```

```python
import pdb
```

```python
x = [1, 2, 3]
y = 2
z = 3

result = y + z
print(result)
pdb.set_trace()   #to stop stack trace type q in ipdb input and it will stops the entire execution.
result2 = x + y
print(y)
print(result2)
```
