Object-Oriented Programming in Python:

Class:
    A class in Python is a blueprint for creating objects
> It defines the attributes and methods that objects of that class will have.
> Classes are a fundamental concept in object oriented programming

Object:
    An object is a concept representing a specific instance of a class.
>It encapsulates data and behaviour that operates on the data.
>Every object has a unique identity, type and value.

Instance:
    Objects are created based on a class which serves the blueprint defining the structure and behaviour of objects.

Attributes:
    Objects hold data through attributes, which are variables specific to each instance.

Identity:
    Each object has a unique identity allowing it to be distinguished from other objects.

Methods:
    Objects can perform actions through methods which are functions defined within the class that operates all the objects attribute.

Encapsulation: 
        Encapsulation in Python means hiding the internal data of a class through setter and
allowing access through getter.(only through methods to protect it from unintended changes).

```python
class Circle():
    pi = 3.14
    def __init__(self, radius):  #global 
        self.radius = radius
        self.area = radius * radius * Circle.pi
    def getCircumference(self):
/        return self.radius * 2 * self.pi  #setter and getter is a combination for encapsulation (data hiding)
    def setRadius(self,newRadius):
        self.radius = newRadius
        self.area = newRadius * newRadius * self.pi
```

```python
c = Circle(10)   
```

    Data binding in Python means assigning data (like values or objects) to variable names so they can be used in a program.
