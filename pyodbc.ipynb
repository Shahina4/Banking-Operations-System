```python
#This line is used to establish a connection between your Python code and a SQL Server database using the pyodbc library.
connection_string = pyodbc.connect("DRIVER={SQL Server};Server=LAPTOP-B34KO8S6\\SQLEXPRESS; Database=CustomerDetails; Trusted_connection=yes")
```

```python
#You're creating a cursor object from your database connection.
cursor = connection_string.cursor() 
#creates a cursor object. The cursor object is a pointer or controller that allows your Python program to perform SQL operations on DB
```

```python
#Execute a SQL query
cursor.execute('SELECT * from Customers')
```

```python
#Retrieve data from the database
row = cursor.fetchall()
```
