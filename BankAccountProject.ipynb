```python
import re
class Bank():
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance
    def balanceAmount(self):
        print(f'Available balance: Rs. {self.balance}')
    def deposit(self, amount):
        if amount>0:
            self.balance += amount
            print(f'Deposit successful, your new balance is $ {self.balance}')
        else:
            print('Invalid deposit amount')
    def withdrawal(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f'Withdrawal successful, your new balance is $ {self.balance}')
        else:
            print('Insufficient/Invalid withdrawal amount')            
class SavingsAccount(Bank):
    def __init__(self, account_number, balance=0, interest_rate=0.01):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate
    def add_interest(self):
        interest = self.balance * self.interest_rate
        self.balance += interest
        print(f'Interest ${int(interest)} added, your new balance is ${int(self.balance)}')

class CurrentAccount(Bank):
    def __init__(self, account_number, balance=0, overdraft_limit=100):
        super().__init__(account_number, balance)
        self.overdraft_limit = overdraft_limit

    def withdrawal(self, amount):
        if amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print(f'Withdrawal successful, your new balance is $ {self.balance}')
        else:
            print('Exceeds balance and overdraft limit.')

def banking():
    print('\n*****Banking Application*****\n')
    accIn = input('Select the account type:\n1. Savings\n2. Current\n\n')
    if accIn == '1' or accIn.lower() == 'savings':
        try:
            acntNumIn = input('Enter account number:')
            pattern = re.search(r'^SA\d{3}', acntNumIn)
            acntNum = pattern.group()
            savings = SavingsAccount(acntNum, 5000)
            print("\n******Menu******\n")
            op = input('1. Balance\n2. Deposit\n3. Withdrawal\n')
            if op == '1' or op.lower() == 'balance':
                savings.balanceAmount()
            elif op == '2' or op.lower() == 'deposit':
                deposit_amount = int(input("Enter Amount:"))
                savings.deposit(deposit_amount)
            elif op == '3' or op.lower() == 'withdrawal':
                withdraw_amount = int(input("Enter Amount:"))
                savings.withdrawal(withdraw_amount)
                savings.add_interest()
            else:
                print('Invalid Input')
        except:
            print("Invalid account number")
            return
    elif accIn == '2' or accIn.lower() == 'current':
        try:
            acntNumIn = input('Enter account number:')
            pattern = re.search(r'^CA\d{3}', acntNumIn)
            acntNum = pattern.group()
            current = CurrentAccount(acntNum, 2000)
            print("\n******Menu******\n")
            op = input('1. Balance\n2. Deposit\n3. Withdrawal\n')
            if op == '1' or op.lower() == 'balance':
                current.balanceAmount()
            elif op == '2' or op.lower() == 'deposit':
                deposit_amount = int(input("Enter Amount:"))
                current.deposit(deposit_amount)
            elif op == '3' or op.lower() == 'withdrawal':
                withdraw_amount = int(input("Enter Amount:"))
                current.withdrawal(withdraw_amount)
            else:
                print('Invalid Input')
        except:
            print("Invalid account number")
            return
    else:
        print('Invalid Selection')

banking()
```
