semantic search : looking for keyword inside files without opening  Ex: resume shortlists, identifica

```python
text = 'My phone number is (555)-555-5555'
```

```python
import re
```

```python
re.search(r'phone', text)
```

```python
re.search(r"(\d{3})-\d{3}-\d{3}", text)
```

```python
match1 = re.search(r"\(\d{3}\)-\d{3}-\d{3}", text)
```

```python
print(match1)
```

```python
match1.span()
```
