Functions are blocks of reusable code that perform a specific task. 
Functions help you write clean, organized, and efficient code.

1. Creating/defining a function - ready to use
2. Function call - Executes when call happens

Syntax:

    #Block of code
    return value     #Ceases/stops the execution and it is optional
}
