```python
num_list =[34,75,25,44,66]

def check_even_list(num_list):
    for number in num_list:
        print('Start')
        if number%2!= 0:
            print('Inside outer if')
            if(number >= 30):
                print('lesser')
            else:
                print('greater')
                continue
        else:
            print('Outside')
        print('End')
    return True

check_even_list(num_list)
```

```python
def check_even_list(num_list):
    for number in num_list:
        if number%2 ==0:
            return True
        else:
            pass
    return False
print("Above function is successful")
```

```python
def check_even_list(num_list):
    even_numbers = []
    for number in num_list:
        if number%2!=0:
            even_numbers.append(number)
        else:
            pass
    return even_numbers
```
