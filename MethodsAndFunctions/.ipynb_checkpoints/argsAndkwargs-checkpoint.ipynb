*args and **kwargs:

1. *args and **kwargs are special syntax in Python used to pass a variable number of arguments to a function.
2. *args is used to pass non-keyword variable length argument list to a function.
3. It collects all postional arguments into a tuple within the function.
4. **kwargs are used to pass keyword variable length argument list to a function.
5. It collects all keyword arguments into a dictionary within a function.
6. IMP: The order of parameters in a function definition is important, positional arguments should come first followed by *args and finally **kwargs.

```python
def new_function(*args): #args saves input as tuple by default
    for item in args:
        print(item)
```

```python
new_function(43,6,32,6,'in','at','#')
```

```python
def my_function(**kwargs): #kwargs saves input as dictionaries by default
    if 'fruit' in kwargs:
        print("My fruit of choice is: {}".format(kwargs['fruit']))
    else:
        print("I did not find any fruit here")
```

```python
my_function(veg='tomato', car='i20', animal='dog', fruit='apple')
```

```python
my_function(vegetable='cucumber')
```

```python
def my_new_function(*args, **kwargs):
    print(args)
    print(kwargs)
    print("I would like to order {} {}".format(args[-1],kwargs['food']))
```

```python
my_new_function(10,20,30,40,425,45,7,1243,47,124, food='biscuit', fruit='apple', chocolate='kitkat')
```

```python
def my_new_function(*args, **kwargs):
    print(args)
    print(kwargs)
    print("I would like to order {} {}".format(args[10],kwargs['food']))
```

```python
my_new_function(10,20,30,40,425,45,7,1243,47,124, food='biscuit', fruit='apple', chocolate='kitkat')
```
