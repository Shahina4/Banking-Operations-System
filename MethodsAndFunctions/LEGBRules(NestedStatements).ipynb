LEGB Rules(Nested Statements):

> L : Local - Names assigned in anywhere within a function and not declared global in that function.

> e : enclosing - Names in the local scope of any and all enclosing function from inner and outer.

> g : global - Names assigned at the top level of a module file are declared global in a function within a file.

> b : built-in - Names pre-assigned in the built-in names module like open, range, etc.

```python
name = 'This is an outer string'

def greet():
    name = 'Learning Again'
    print('Hi: '+name)

    def hello():
    
        print('Hello: '+name)

        def demo():
            name = 'Learning'
            print('demo: '+name)
        demo()
    hello()
print(name)
```

```python
greet()
```

```python
name = 'This is an outer string'   #global

def greet():
    name = 'Learning Again'  #local
    print('Hi: '+name)

    def hello(): 
        #name='New Learning'
        print('Hello: '+name)  #enclosing to greet() name

        def demo():
            #name = 'Learning'
            print('demo: '+name)  #enclosing to greet() name
        demo()
    hello()
print(name)
```
